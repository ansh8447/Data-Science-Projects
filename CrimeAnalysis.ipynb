```python
# For bradbury question (34.050536, -118.247861)

# D = R sqrt ( square(delta_phi) + square(cos(phi_mean)*delta_lambda) )
# where delta_phi = lat2 - lat1 (in radians)
# delta_lambda = lon2 - lon1 (in radians)
# phi_mean = (lat1 + lat2)/2 (in radians)
# To convert in radian, we can multiply by (pi/180)
# R = 6371 km

# Add a column to the dataset which will have 1 if the location is within 2km else will have 0. 
# Calculate the sum of the added column.

lat1 = 34.050536
long1 = -118.247861


# Importing re package for using regular expressions 
import re 
import math
  
# Function to clean the names 
def Clean_names(locations): 
    # Search for opening bracket in the name followed by 
    # any characters repeated any number of times 
    if re.search('\(.*', locations): 
  
        # Extract the position of beginning of pattern 
        pos_s = re.search('\(.*', locations).start()        
        pos_e = re.search('\(.*', locations).end() 
          
        # return the cleaned name 
        return locations[pos_s+1:pos_e-1] 
  
    else: 
        # if clean up needed return the same name 
        return locations 
          
# Updated the city columns 
#complete_df_cols_wthout_na['Location'] 
complete_df_cols_wthout_na['Location'] = complete_df_cols_wthout_na['Location'].apply(Clean_names) 
  
# Print the updated dataframe 
complete_df_cols_wthout_na[['lat', 'long']] = complete_df_cols_wthout_na.Location.str.split(', ',expand=True)
complete_df_cols_wthout_na['lat'] = complete_df_cols_wthout_na['lat'].astype(float)
complete_df_cols_wthout_na['long'] = complete_df_cols_wthout_na['long'].astype(float)


complete_df_cols_wthout_na['delta_phi'] = complete_df_cols_wthout_na['lat'].apply(lambda x : (x - 34.050536)*(3.14/180))
complete_df_cols_wthout_na['delta_lambda'] = complete_df_cols_wthout_na['long'].apply(lambda x : (x + 118.247861)*(3.14/180))

#Calculating cosine of delta_mean
complete_df_cols_wthout_na['delta_mean'] = complete_df_cols_wthout_na['lat'].apply(lambda x : (x + 34.050536)/2)


location_quest = complete_df_cols_wthout_na[['lat','long','delta_phi', 'delta_lambda']]
#REmoving lats with value 0
location_quest = location_quest[location_quest['lat'] != 0]

location_quest['delta_mean'] = location_quest['lat'].apply(lambda x: math.cos((x+34.050536)/2))

location_quest['sq_delta_phi'] = location_quest['delta_phi'].apply(lambda x: x**2)
location_quest['sq_delta_lambda'] = location_quest['delta_lambda'] * location_quest['delta_mean']
location_quest['sq_delta_lambda'] = location_quest['sq_delta_lambda']**2

location_quest['sum_squares'] = location_quest['sq_delta_lambda'] + location_quest['sq_delta_phi']
location_quest['total_distance'] = location_quest['sum_squares'].apply(lambda x: math.sqrt(x)*6371)

#Crimes within distance of 2km
location_quest['two_km'] = location_quest['total_distance'].apply(lambda x: 1 if (x <= 2.0) else 0) 
location_quest['two_km'].sum()
```

```python
#removing outliers

pico_df = complete_df_cols_wthout_na[complete_df_cols_wthout_na['Address'].str.contains('PICO')]
pico_lat_mean = pico_df['lat'].mean()
pico_long_mean = pico_df['long'].mean()
pico_lat_std = pico_df['lat'].std()
pico_long_std = pico_df['long'].std()

lat_idx = pico_df['lat'] <= pico_lat_mean + (pico_lat_std * 2)
long_idx = pico_df['long'] <= pico_long_mean + (pico_long_std * 2)
pico_df = pico_df[lat_idx & long_idx]
lat_idx_2 = pico_df['lat'] >= pico_lat_mean - (pico_lat_std * 2)
long_idx_2 = pico_df['long'] >= pico_long_mean - (pico_long_std * 2)
pico_df = pico_df[lat_idx_2 & long_idx_2]
```

```python
pico_df = pico_df.drop(['delta_phi','delta_lambda','delta_mean'], axis=1)
```

```python
#For most eastern and most western, we'll calculate min and max of longitudes
long_min = pico_df['long'].min()
long_max = pico_df['long'].max()
new_df = pico_df.loc[(pico_df['long']==long_min) | (pico_df['long']==long_max)]
new_df = new_df.iloc[[0,5]]

lat1 = new_df.iloc[0].lat
long1 = new_df.iloc[0].long
lat2 = new_df.iloc[1].lat
long2 = new_df.iloc[1].long

delta_phi_sq = ((lat1-lat2)**2)*(3.14/180)
delta_lambda = (long1-long2)*(3.14/180) 
cos_phi_m = (math.cos((lat1 + lat2)/2))*(3.14/180)
distance = 6371* math.sqrt(delta_phi_sq + ((cos_phi_m * delta_lambda)**2))

arrests = pico_df['Location'].count()
per_km = arrests/distance
per_km
```
